```python

#   DecodeLabs | Batch 2026 | Project 1
#   Rule-Based AI Chatbot
#   Author  : Mahidi Hasan Mithun

# ────────────────────────────────────────────────────────────
#  PHASE 0 ── IMPORTS  (only stdlib needed)
# ────────────────────────────────────────────────────────────
import time
import random

# ────────────────────────────────────────────────────────────
#  PHASE 1 ── KNOWLEDGE BASE  (Hash-Map / Dictionary)
# ────────────────────────────────────────────────────────────

KNOWLEDGE_BASE = {

    # ── Greetings ──────────────────────────────────────────
    "hello"       : ["Hello! I'm DecoBot. How can I assist you today?",
                     "Hey there! DecoBot online. What do you need?",
                     "Hi! Welcome to DecodeLabs. How may I help?"],

    "hi"          : ["Hi! Great to see you. Ask me anything.",
                     "Hey! DecoBot here — ready to help."],

    "hey"         : ["Hey! What's on your mind?",
                     "Hello! Fire away with your question."],

    "good morning": ["Good morning! Hope you're ready to build something amazing today!",
                     "Morning! Let's write some great code today."],

    "good evening": ["Good evening! Still coding? Let's go!",
                     "Good evening! DecoBot is here for you."],

    # ── Farewells ──────────────────────────────────────────
    "bye"         : ["Goodbye! Keep building. See you soon!",
                     "Bye! You're one step closer to being an AI Engineer."],

    "goodbye"     : ["Goodbye! Remember: every rule you write is a step toward mastery.",
                     "See you! Great work today."],

    "see you"     : ["See you! Keep experimenting.",
                     "Take care! Come back anytime."],

    # ── How Are You ────────────────────────────────────────
    "how are you" : ["I'm a rule-based system — I don't feel things, but my logic is running perfectly!",
                     "All systems operational! Thanks for asking.",
                     "Running at 100% deterministic efficiency!"],

    "what's up"   : ["Not much — just processing rules at O(1) speed. You?",
                     "All good in the logic engine!"],

    # ── About the Bot ─────────────────────────────────────
    "who are you" : ["I'm DecoBot — a Rule-Based AI Chatbot built for DecodeLabs Project 1.",
                     "DecoBot! A deterministic chatbot powered by pure if-else logic."],

    "what are you": ["I'm a rule-based chatbot. I match your input to a knowledge base and return a response.",
                     "A rule-based AI — no neural networks, just clean deterministic logic."],

    "what can you do": ["I can answer predefined questions, greet you, tell you about AI, and help you understand rule-based systems!",
                        "I respond to specific intents using a dictionary. Ask me about AI, Python, or DecodeLabs!"],

    # ── About AI ──────────────────────────────────────────
    "what is ai"  : ["AI stands for Artificial Intelligence — machines that simulate human-like decision-making. "
                     "I'm a simple example: rule-based AI!",
                     "Artificial Intelligence is the science of making machines smart. "
                     "It ranges from simple rule systems (like me!) to complex neural networks."],

    "what is machine learning": [
                     "Machine Learning is a subset of AI where systems LEARN from data instead of following hard-coded rules. "
                     "Unlike me — I follow explicit rules!",
                     "ML lets computers find patterns in data automatically. "
                     "The next project at DecodeLabs will take you closer to that!"],

    "what is deep learning": [
                     "Deep Learning uses multi-layer neural networks to learn complex patterns — "
                     "like image recognition and language translation.",
                     "Deep Learning is a subset of ML using artificial neural networks with many layers. "
                     "It's what powers ChatGPT and similar models!"],

    "what is nlp" : ["NLP stands for Natural Language Processing — "
                     "it's how computers understand and generate human language. "
                     "I use a very basic form of NLP: exact keyword matching!",
                     "Natural Language Processing helps machines read, understand, and respond to text. "
                     "Transformers like BERT and GPT are state-of-the-art NLP models."],

    # ── About Python ──────────────────────────────────────
    "what is python": ["Python is a high-level, easy-to-read programming language. "
                       "It's the #1 choice for AI and Data Science!",
                       "Python is the language I'm written in! It's powerful, readable, and has amazing libraries for AI."],

    "what is a dictionary": [
                     "A Python dictionary is a key-value data structure with O(1) lookup time. "
                     "My entire brain IS a dictionary!",
                     "Dictionaries store data as key:value pairs. "
                     "They're much faster than if-elif ladders for chatbot logic — as shown in your slides!"],

    "what is a loop": ["A loop repeats a block of code. I use a 'while True' loop as my heartbeat — "
                       "I keep running until you say 'exit'!",
                       "Loops are control-flow structures. My infinite loop keeps me alive and listening."],

    # ── About DecodeLabs / Project ────────────────────────
    "what is decodelabs": ["DecodeLabs is an industrial training platform that gives interns real-world AI projects. "
                            "You're currently completing Project 1!",
                            "DecodeLabs helps aspiring engineers build real portfolios through hands-on AI projects."],

    "what is project 1": ["Project 1 is the Rule-Based AI Chatbot milestone. "
                           "You build a chatbot using control flow, dictionaries, and loops — "
                           "mastering the logic engine before moving to probabilistic AI.",
                           "Project 1 = building ME! A rule-based chatbot using Python dictionaries and while loops."],

    "what is rule based ai": [
                     "Rule-based AI uses explicit IF-ELSE or dictionary rules to make decisions. "
                     "It's deterministic (same input → same output), fully traceable, and zero hallucination risk!",
                     "Rule-based AI is the 'white box' — every decision is explainable. "
                     "I'm a perfect example: your input maps directly to a response."],

    "what is ipo model": ["The IPO model stands for Input → Process → Output. "
                          "For me: Input = your text, Process = dictionary lookup, Output = my response!",
                          "IPO (Input-Process-Output) is the foundational blueprint for transparent AI systems, "
                          "as described in your DecodeLabs slides."],

    "what are guardrails": ["AI guardrails are rule-based filters that sit on top of LLMs — "
                             "they block harmful inputs and ensure safe, compliant outputs. "
                             "Frameworks like NVIDIA NeMo and Llama Guard use this approach!",
                             "Guardrails = the deterministic control layer around probabilistic AI. "
                             "You're building the foundation of that right now!"],

    # ── Fun / Personality ──────────────────────────────────
    "tell me a joke": ["Why do programmers prefer dark mode? Because light attracts bugs! 🐛",
                       "Why did the AI go to school? To improve its 'neural' network! 😄",
                       "I told my dictionary a joke. It had no response for that key. 😂"],

    "what is your name": ["I'm DecoBot — your DecodeLabs Project 1 AI assistant!",
                           "Call me DecoBot. Built with Python, powered by rules."],

    "thank you"   : ["You're welcome! Keep coding!",
                     "Happy to help! You're going to be a great AI Engineer.",
                     "Anytime! That's what rule-based logic is for."],

    "thanks"      : ["No problem! Keep building.",
                     "You're welcome! Stay curious."],

    "help"        : ["Sure! You can ask me about: AI, Machine Learning, Deep Learning, NLP, Python, "
                     "DecodeLabs, Project 1, guardrails, the IPO model, or just chat. "
                     "Type 'exit' to quit.",
                     "Try asking: 'what is AI', 'what is a dictionary', 'tell me a joke', or 'what is rule based AI'."],
}

# ────────────────────────────────────────────────────────────
#  PHASE 2 ── SANITIZATION ENGINE
# ────────────────────────────────────────────────────────────

def sanitize(raw_input: str) -> str:
    """
    Normalize raw user input for consistent dictionary lookup.
    Pipeline: strip whitespace → lowercase → remove trailing punctuation
    """
    clean = raw_input.strip()        # remove leading/trailing spaces
    clean = clean.lower()            # normalize case  ("Hello" → "hello")
    clean = clean.rstrip("?!.,;:")   # remove trailing punctuation
    clean = " ".join(clean.split())  # collapse multiple internal spaces
    return clean


# ────────────────────────────────────────────────────────────
#  PHASE 3 ── INTENT MATCHING ENGINE  (The Logic Skeleton)
# ────────────────────────────────────────────────────────────

FALLBACK_RESPONSES = [
    "I don't understand that yet. Try asking about AI, Python, or DecodeLabs!",
    "Hmm, that's outside my current knowledge base. Type 'help' to see what I can do.",
    "I'm a rule-based bot — I only know what I've been taught. Ask me something else!",
    "No rule found for that. I'm learning... well, actually I'm not. I'm deterministic. Try 'help'!",
]

EXIT_COMMANDS = {"exit", "quit", "q", "stop", "close", "bye bye", "end"}


def get_response(clean_input: str) -> str | None:
    """
    Match clean input to a response using a 3-tier strategy.

    Returns:
        str  → the bot's reply
        None → signals EXIT
    """

    # ── Tier 0: Exit Check ────────────────────────────────
    if clean_input in EXIT_COMMANDS:
        return None  # None = kill signal for the main loop

    # ── Tier 1: Exact Match  O(1) ─────────────────────────
    if clean_input in KNOWLEDGE_BASE:
        return random.choice(KNOWLEDGE_BASE[clean_input])

    # ── Tier 2: Keyword / Partial Match  O(n) fallback ────
    # Scan all keys to find a partial intent match
    for key in KNOWLEDGE_BASE:
        if key in clean_input or clean_input in key:
            return random.choice(KNOWLEDGE_BASE[key])

    # ── Tier 3: Fallback ──────────────────────────────────
    return random.choice(FALLBACK_RESPONSES)


# ────────────────────────────────────────────────────────────
#  PHASE 4 ── OUTPUT ENGINE  (Pretty Printing)
# ────────────────────────────────────────────────────────────

def bot_print(message: str, delay: float = 0.03):
    """
    Simulate typing effect — prints character by character.
    Makes the chatbot feel more alive.
    """
    print("🤖 DecoBot: ", end="", flush=True)
    for char in message:
        print(char, end="", flush=True)
        time.sleep(delay)
    print()  # newline after response


def print_banner():
    """Display the startup banner."""
    banner = """
────────────────────────────────────────────────────────────
          DecodeLabs | Batch 2026 | Project 1
            RULE-BASED AI CHATBOT  🤖

       Type 'help' to see what I know.
       Type 'exit' to shut me down.
────────────────────────────────────────────────────────────
    """
    print(banner)


def print_divider():
    print("─" * 60)


# ────────────────────────────────────────────────────────────
#  PHASE 5 ── THE MAIN LOOP  (The Heartbeat)
# ────────────────────────────────────────────────────────────

def run_chatbot():
    """
    Main chatbot loop — the organism stays alive until kill command.
    Implements the IPO model:
        Input  → sanitize()
        Process→ get_response()
        Output → bot_print()
    """
    print_banner()
    bot_print("Hello! I'm DecoBot, your DecodeLabs AI assistant. How can I help you today?")
    print_divider()

    conversation_count = 0  # track how many exchanges happened

    # ── THE INFINITE LOOP ──────────────────────────────────
    while True:

        # ── INPUT PHASE ─────────────────────────────────
        try:
            raw_input_text = input("\n👤 You: ")
        except (EOFError, KeyboardInterrupt):
            # Graceful exit on Ctrl+C or Colab interruption
            print()
            bot_print("Session interrupted. Goodbye!")
            break

        # Guard: ignore empty input
        if not raw_input_text.strip():
            bot_print("Please type something! (or 'exit' to quit)")
            continue

        # ── PROCESS PHASE 1: SANITIZE ───────────────────
        clean = sanitize(raw_input_text)

        # ── PROCESS PHASE 2: INTENT MATCH ───────────────
        response = get_response(clean)

        # ── EXIT STRATEGY: Clean break command ──────────
        if response is None:
            print_divider()
            bot_print(f"Goodbye! We had {conversation_count} exchanges. "
                      "Keep building — your AI journey has just begun! 🚀")
            print_divider()
            break

        # ── OUTPUT PHASE ────────────────────────────────
        print()
        bot_print(response)
        conversation_count += 1

        # Optional: show exchange counter every 5 turns
        if conversation_count % 5 == 0:
            print(f"   [ℹ️  {conversation_count} exchanges so far — you're on a roll!]")


# ────────────────────────────────────────────────────────────
#  ENTRY POINT
# ────────────────────────────────────────────────────────────

if __name__ == "__main__":
    run_chatbot()
```
